```python
from google.colab import files
uploaded = files.upload()

import io
# Assuming the uploaded file is the model joblib file
model_filename = next(iter(uploaded))
```

```python
import joblib

# Load the model
model = joblib.load(model_filename)

# Prediction function
def predict(text: str):
    prediction = model.predict([text])
    return prediction[0]

# Test the prediction function with a sample text
sample_text = "good job!"
print("Prediction:", predict(sample_text))
```

```python
# Now testing on test tweets
test_texts = [
    "Absolutely loved the movie!",
    "Worst experience ever.",
    "It was okay, nothing special.",
    "I think it was quite impressive, will watch again!",
    "This is terrible. Not what I expected at all.",
    "What a fantastic performance!",
    "Not good, not bad, just meh.",
    "The service was bad but the food was fantastic!"
]

# Iterate through the test texts and print predictions
for text in test_texts:
    print(f"Text: {text}\nPrediction: {predict(text)}\n")
```
